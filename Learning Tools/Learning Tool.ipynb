<h3><b>Different Types of Classifiers</b></h3>
<p><ul>
<li>K-Nearest Neighbors
    <p>K-Nearest Neighbors operates by checking the distance from some test example to the known values of some training example. The group of data points/class that would give the smallest distance between the training points and the testing point is the class that is selected.
    </p><pre>from sklearn.neighbors import KNeighborsClassifier</pre></li><br>
<li>Decision Trees
    <p>A Decision Tree Classifier functions by breaking down a dataset into smaller and smaller subsets based on different criteria. Different sorting criteria will be used to divide the dataset, with the number of examples getting smaller with every division.
    </p><pre>from sklearn.tree import DecisionTreeClassifier</pre></li><br>
<li>Naive Bayes
    <p>A Naive Bayes Classifier determines the probability that an example belongs to some class, calculating the probability that an event will occur given that some input event has occurred. When it does this calculation it is assumed that all the predictors of a class have the same effect on the outcome, that the predictors are independent.
    </p><pre>from sklearn.naive_bayes import GaussianNB</pre></li><br>
<li>Linear Discriminant Analysis
    <p>Linear Discriminant Analysis works by reducing the dimensionality of the dataset, projecting all of the data points into a line. Then it combines these points into classes based on their distance from a chosen point or centroid.
    </p><pre>from sklearn.discriminant_analysis import LinearDiscriminantAnalysis</pre></li><br>
<li>Support Vector Machines
    <p>Support Vector Machines work by drawing a line between the different clusters of data points to group them into classes. Points on one side of the line will be one class and points on the other side belong to another class.
    </p><pre>from sklearn.svm import SVC</pre></li><br>
<li>Logistic Regression
    <p>Logistic Regression outputs predictions about test data points on a binary scale, zero or one. If the value of something is 0.5 or above, it is classified as belonging to class 1, while below 0.5 if is classified as belonging to 0. Each of the features also has a label of only 0 or 1. Logistic regression is a linear classifier and therefore used when there is some sort of linear relationship between the data.
    </p><pre>from sklearn.linear_model import LogisticRegression</pre></li><br>
</ul></p>
